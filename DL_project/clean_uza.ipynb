```python
from sklearn.datasets import load_files

news = load_files('.\\baza_mini', encoding='utf-8')

posts = news.data
categories = news.target

labels = news.target_names

import re, string

clean_categories = []
for category in categories:    
    clean_categories.append(labels[category])
    
    
clean_posts = []
for post in posts:    
    post = post.lower()
    post = "".join([char for char in post if char not in '!"#$%&\'()*+.,-/:;<=>?@[\\]^_`{|}~–“”ʼ«»'])
    post = re.sub('[0-9]+', '', post)
    post = re.sub('(\r\n|\r|\n)', ' ', post)   
    clean_posts.append(post)
    
```

```python
print(clean_posts[1])
print(clean_categories[1])
```

```python
import pandas as pd  
df = pd.DataFrame({'category': clean_categories, 'post': clean_posts}) 
df.to_csv('clean_uza_mini.csv')
```
